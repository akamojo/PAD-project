## Regression Analysis



```python
# Numpy for data management
import numpy as np

# Pandas also for data management
import pandas as pd

# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline

# Seaborn for plotting and styling
import seaborn as sns
sns.set_style("darkgrid")

# Stats modules
import statsmodels.api as sm
import scipy.stats as stats

# Auxiliar libraries
from sklearn.metrics import r2_score
```

```python
# load the text file
from numpy import genfromtxt
data = genfromtxt('tripadvisor_review.csv', delimiter=',')
data = data[1:, 1:]
data.shape
```

```python
x = np.reshape(data[:,5], (-1, 1)) # Resorts
y = data[:,4] # Museums

x.shape
```

```python
ax = sns.regplot(x=x, y=y, fit_reg=False, scatter_kws={'alpha':0.5})
ax.set(xlabel='Resorts', 
       ylabel='Museums');
```

Originally, in the dataset *'Travel Reviews'* we chose there was no connection between the average feedback on resorts and museums, so we can say that these two variables are independent. On the other hand, as we are looking at the scatter plot, we can notice the following *'linear‚Äêlike'* dependence: the better resorts are the better are also museums. We can also see that resorts are slightly better then museums. Generally the average feedback on resorts is concentrated between 1 and 2.5 and as for feedback on museums, it is concentrated between 0.5 and 1.5.

__Linear Regression model without intercept__

```python
beta = np.linalg.lstsq(x, y)[0]
print('Estimated coefficient (slope):', beta[0])
```

```python
predictions = x * beta # making use of numpy's broadcast
predictions_withouth_intercept = predictions

ax = sns.regplot(x=x, y=y, fit_reg=False, scatter_kws={'alpha':0.5})
ax.set(xlabel='Museums', 
       ylabel='Resorts',
       title='Linear Regression Relation bt. X & Y');
plt.plot(x, predictions) # overlay a line plot over a scatter plot 
plt.show()
```

```python
constant_term = np.ones(len(x)) 
x = np.c_[constant_term, x]

print(x[:5])
```

__Linear Regression model with intercept__

```python
beta = np.linalg.lstsq(x, y)[0]
print('Estimated coefficients:', beta)

# to use the dot product function we have to state that the coefficients vector is of the shape (2,1)
print('Old shape of coefficients array:',beta.shape)
beta = np.reshape(beta,(-1,1))
print('New shape of coefficients array:',beta.shape)
```

```python
from numpy.linalg import inv
from numpy.linalg import det
from numpy import dot

# compute new predictions
predictions_with_intercept = dot(x,beta) 

ax = sns.regplot(x=x[:,1], y=y, fit_reg=False, scatter_kws={'alpha':0.5})
ax.set(xlabel='Museums', 
       ylabel='Resorts',
       title='Linear Regression Relation bt. X & Y');
plt.plot(x[:,1], predictions_with_intercept) # overlay a line plot over a scatter plot 
plt.show()
```

```python
def ss_total(y):
    y_av = np.average(y)
    res = 0
    for i in y:
        res += (i - y_av)**2
    return res

def ss_reg(pred, y):
    y_av = np.average(y)
    res = 0
    for i in pred:
        res += (i - y_av)**2
    return res[0]

def ss_res(pred, y):
    res = 0
    for i in range(len(pred)):
        res += (y[i] - pred[i])**2
    return res[0]

sstotal = ss_total(y)
ss_total_using_variance = (len(y) - 1 ) * np.var(y, ddof=1) 
ssreg = ss_reg(predictions_with_intercept, y)
ssres = ss_res(predictions_with_intercept, y)
rsq_with_intercept = 1 - (ssres / sstotal)

print('SStotal:', round(sstotal,4))
print('SSreg:', round(ssreg,4))
print('SSres:', round(ssres,4))
print('Coefficient of Determinacy, model with intercept:', np.round(rsq_with_intercept,2))
```

__Normal probability plot of the standarized residuals from regression__

```python
fig, axs = plt.subplots(ncols=2, # Set the number of columns to split the figure in
                        figsize=(16, 6)) # size of the whole figure

predictions = dot(x,beta)
residuals = y - predictions[:,0]
degrees_of_freedom = len(y) - x.shape[1]

sm.qqplot(residuals, 
          stats.t,
          distargs=(degrees_of_freedom,), 
          line='q', 
          ax=axs[0])
axs[0].set_title(label = 'Q-Q plot')
axs[0].set_xlim(-3.5,3.5)


residuals_standard_error = np.std(residuals)
mean_independent_variable = np.mean(x[:,1])

pom1 = 0
for i in x[:,1]:
    pom1 += (i - mean_independent_variable)**2
    
leverages = np.array([(1/len(y) + (i - mean_independent_variable)**2 / pom1) for i in x[:,1]])
sid = residuals_standard_error * np.sqrt(np.full(leverages.shape, 1) - leverages)
standardized_rediduals = residuals / sid

sns.regplot(x=predictions[:,0], 
            y=standardized_rediduals, 
            ax=axs[1],
            lowess=True, 
            scatter_kws={'alpha':0.3}, 
            line_kws={"color":"r","alpha":0.4,"lw":2})
plt.plot(np.arange(len(standardized_rediduals)), [0]*len(standardized_rediduals), 'r-')
axs[1].set_ylim(-3.5,3.5)
axs[1].set_xlim(0, 2)
axs[1].set(ylabel='Standardized Residuals', 
           xlabel='Fitted values');
axs[1].set_title(label = 'Standardized Residuals vs Fitted values');
```

Looking at the Quantile-Quantile plot (on the left) of quantiles of our residuals' distribution against quantiles of standard normal distribution we can notice that it is is right-skewed and therefore, we decided to check if applying log transformation to the data would improve our results. On the other hand, looking on the plot of standardized residuals against fits (Predicted Values) we can notice that it displays *'funnel'* pattern, which violates constant variance assumption. Residual variance lower for smaller values of x, where variance increases as x-values increase.

```python
import scipy
print("Mean:", np.mean(residuals))
print("Skewness:", scipy.stats.skew(residuals))
```

The value of skewness is bigger than 0 so it confirms the fact that the distribution of residuals is right-skewed.

__Population regression equation for the model__

$y_i = B_0 + B_1 x_i + e_i,  i = 1,...,n $

In the population regression equation of our model, the value of parameter $B_0$ is the true value of intercept that we want to estimate. The $B_1$ parameter is the true value of slope coefficient.

__Test of statistical hypothesis for determining whether a linear relationship exists between average user feedback on Museums and Resorts__

- H0: asserts $B_1 = 0$ (no linear relationship exists)
- H1: asserts $B_1 ‚â† 0$ (linear relationship exists)

```python
beta = np.linalg.lstsq(x, y)[0]
beta
```

```python
beta = np.reshape(beta,(-1,1))
predictions = dot(x,beta)

residuals = y - predictions[:,0]
sigma_degrees_of_freedom = len(x) - x.shape[1]
see = sum(residuals ** 2) / sigma_degrees_of_freedom
sigma_sqr = np.std(y)

variance_covmatrix = np.linalg.inv(x.T.dot(x))
coeff_stde = see * variance_covmatrix
coeff_stde = np.sqrt(coeff_stde[coeff_stde > 0])

print('Coefficients\' Standard Errors:',coeff_stde)
```

```python
expected_value = 0.0  
ttest = (beta[1] - expected_value) / coeff_stde[1]

#Degrees of freedom
df = x.shape[0] - 2
pval = stats.t.sf(np.abs(ttest), df)*2  # two-sided pvalue = Prob(abs(t)>tt)

print('Test statistc for coefficient estimate B_1:',ttest,'| P-value:', np.round(pval,100))
```

Because p-value ~ 0.000, H0 (that asserts $B_1 = 0$) is rejected, so it indicates that a linear relationship exists between average feedback on museums and resorts.

__Construction and interpretation of 95% confidence interval for the unknown true slope of the regression line__

```python
df = len(x) - x.shape[1]
alpha = 0.05
one_minus_alpha = 1 - alpha
t_student_values = stats.t.ppf( [alpha,one_minus_alpha], df)
print(beta[1] + t_student_values * coeff_stde[1])
```

Taking into consideration obtained interval, we can say that 95% confident true slope of regression line lies between 0.44 and 0.51.

__Construct and interpret a 95% confidence interval for the mean of the y-variable at a fixed value od other variable__

```python
x_mean = np.mean(x[:,1])
conf_interval = t_student_values * see * np.sqrt((1 / len(x)) + (2.22 - x_mean)**2 / sum((x[:,1] - x_mean)**2))
y_predicted = beta[1] * 2.22 + beta[0]
print(y_predicted + conf_interval)
```

__Construct and interpret a 95% confidence interval for the mean of the y-variable at a fixed value od other variable__

```python
conf_interval = t_student_values * see * np.sqrt(1 + (1 / len(x)) + (2.22 - x_mean)**2 / sum((x[:,1] - x_mean)**2))
y_predicted = beta[1] * 2.22 + beta[0]
print(y_predicted + conf_interval)
```

Looking at the confidence intervals which we obtained, we can notice that the prediction interval for value of randomly chosen y given x is wider. It is harder to predict the particular value of y given x so we have to broaden the interval. The reason for that is the fact that random variable x has standard deviation sigma while sample mean x-bar has standard deviation sigma/n.
