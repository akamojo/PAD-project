__Visualization of the data over chosen features using two types of normalization__

```python
def visualizeSVD(center_data):
    u, s, vh = np.linalg.svd(center_data, full_matrices=True)
    project_data_featureloading = np.dot(center_data, vh.T) # Transpose feature loadings 'vh'
    v_norm = np.dot(vh.T,vh)
    Us_eig = np.dot(u[:,0:3], np.diag(s[0:3]))
    return Us_eig
```

// maybe PCA visualisation?

```python
def explained_variance_plot(x_norm):
    ## Covariance matrix of the normalized dataset
    covmatrix = np.cov(x_norm.T)
    
    ## Obtain the eingenvalues and eingenvectors of covariance matrix
    # The eigenvalues are not necessarily ordered
    e, v = np.linalg.eig(covmatrix)

    totVar = sum(e)       #Total variability of the Data
    var_exp = [(i / totVar)*100 for i in sorted(e, reverse=True)]
    cum_var_exp = np.cumsum(var_exp)

    trace1 = Bar(
            x=['PC %s' %i for i in range(1,5)],
            y=var_exp,
            name='Exp Var',
            showlegend=False)

    trace2 = Scatter(
            x=['PC %s' %i for i in range(1,5)], 
            y=cum_var_exp,
            name='Cum. Exp Var')

    data = Data([trace1, trace2])

    layout=Layout(
            yaxis=YAxis(title='Explained variance in percent'),
            title='Explained variance by different principal components')

    fig = Figure(data=data, layout=layout)
    plot(fig)
```

```python
row_mean = np.mean(data, axis=0)
center_data = data - row_mean
print(center_data)
```

__Computation and interpretation of a hidden factor behind the selected features__

```python
X = data.copy()

for i in range(3):
    X[:,i] = (X[:,i] / max(X[:,i])) * 100

Z, Mu, C = np.linalg.svd(X, full_matrices=True)
Z = - Z[:, 0]
Mu = Mu[0]
C = - C[0]
print(Z, Mu, C)
```

```python
alpha = 100 / sum(C * 100)
hidden_factor_coeff = C * alpha
print(hidden_factor_coeff)
```

Therefore, our final equation would be:
$Z = 0.16 * Restaurants + 0.31 * Museums + 0.53 * Resorts$

As we can see, weight of Average feedback on Resorts is the highest and is increased in the comparision to the mean, so it has the biggest contribution to the hidden factor. On the other hand, the weight of Average feedback on Restaurants and Museums is lowered in comparision to the mean.

```python
hidden_factor = []

for i in range(len(X)):
    hidden_factor.append(np.dot(X[i], hidden_factor_coeff))

print(np.array(hidden_factor))
```
